```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Antigüedad de saldos 17.03.2023.xlsx')
df
```

```python
#Verificamos información del DataFrame
df.info()
```

ITERACIÓN

**Iterar columnas directamente del dataframe**

```python
# Iterar sobre la columna 'nombre'
for nombre in df['NOMBRE']:
    print(nombre)
```

***USANDO NUMPY***

**Iterar columnas tipo STRING**

```python
# Iterar sobre una columna tipo string: 'NOMBRE' con Numpy
nombres = df['NOMBRE'].to_numpy() #Primero convertimos la columna en un array

# Iterar sobre el array
for nombre in np.nditer(nombres, flags=['refs_ok']): #Usamos flags=['refs_ok'] si la columna contiene strings (objetos) en lugar de números
    print(nombre.item())  # item() lo convierte a string nativo de Python
```

**Iterar columnas tipo NUMÉRICO**

```python
# Iterar sobre una columna tipo numérico: 'MONTO ADEUDADO' con Numpy
monto_adeudado = df['MONTO ADEUDADO'].to_numpy() # Convertir la columna 'MONTO ADEUDADO' a array de NumPy

# Iterar sobre el array
for monto_adeudado in np.nditer(monto_adeudado):
    print(monto_adeudado)  # monto_adeudado es directamente un número (int o float)
```

**Iterar Filas**

```python
# Iterar sobre cada fila del dataframe
for index, row in df.iterrows():
    print(f"Fila {index}:")
    print(f"NOMBRE: {row['NOMBRE']}")
    print(f"MONTO ADEUDADO: {row['MONTO ADEUDADO']}")
    print("-----")
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
filtro7=df[(df["MONTO ADEUDADO"] < 6000) & (df["FECHA_FACTURA"] == "2023-03-14")]
filtro7
```

```python
#Filtro o
filtro8=df[(df["MONTO ADEUDADO"] < 6000) | (df["FECHA_FACTURA"] == "2023-03-14")]
filtro8
```

```python
#Filtro not
filtro9=df[~(df["MONTO ADEUDADO"] >= 6000)]
filtro9
```
