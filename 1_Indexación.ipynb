```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertimos el dataset en dataframe
df= pd.read_csv("titanic.csv")
df
```

```python
#Verificamos información del DataFrame
df.info()
```

***Indexación de un array 1D a partir de un array 2D***

```python
# Acceder a una Fila única (Array 1D)
fila0 = df.iloc[0]
fila0
```

```python
# Acceder a una Columna única (Array 1D)
columna1 = df['survived']
columna1
```

***Indexación de un elemento a partir de un array 1D***

```python
# Acceder a un elemento de la Fila (Array 1D)
Elemento4 = fila0.iloc[[4]]
Elemento4
```

```python
# Acceder a un elemento de la Columna (Array 1D)
Elemento4 = columna1.iloc[[4]]
Elemento4
```

***Indexación de múltiples arrays 1D a partir de un array 2D***

```python
# Acceder a más de una Fila (Array 2D)
filas1 = df.iloc[[0, 8, 50]] #No consecutivas
filas2 = df.iloc[0:10] #Consecutivas
filas3 = df.iloc[[0] + list(range(10, 21)) + [50]] #Combinadas
filas1
```

```python
# Acceder a más de una Columna (Array 2D)
columnas1 = df[['survived', 'age']] #pueden ser consecutivas o no consecutivas
columnas1
```

***Indexación de múltiples arrays 1D a partir de un array 2D combinando FILAS VS COLUMNAS**

```python
#Filas consecutivas y Todas las columnas
df1= df.iloc[3:10, :]       
df1
```

```python
#Filas no consecutivas y Todas las columnas
df2= df.iloc[ [8, 24, 100] , : ] 
df2
```

```python
#Filas combinadas y Todas las columnas
df3= df.iloc[ [0] + list(range(10, 13)) + [50] , : ] 
df3
```

```python
#Columnas consecutivas y Todas las columnas
df4= df.iloc[:, 2:5]       
df4
```

```python
#Convertir array a CSV
df4.to_csv("Entregable1.csv")
```
